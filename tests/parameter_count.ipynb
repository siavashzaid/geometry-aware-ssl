```python
import sys
import torch

sys.path.append('/mnt/data/zaid/projects/geometry-aware-ssl')

from configs.final_config import final_config
from src.training import build_model
from baseline.model import EigmodeTransformer


def count_parameters(model):
    total     = sum(p.numel() for p in model.parameters())
    trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"  Total:     {total:,}")
    print(f"  Trainable: {trainable:,}")


if __name__ == "__main__":

    # --- Baseline ---
    baseline = EigmodeTransformer(
        nchannels    = 64,
        num_layers   = 12,
        num_heads    = 8,
        dropout_rate = 0.1,
    )
    print("EigmodeTransformer (baseline):")
    count_parameters(baseline)

    # --- GNN model ---
    gnn = build_model(final_config)
    print("\nMPNNTransformer (ours):")
    count_parameters(gnn)
```
